```python
import numpy as np
import matplotlib.pyplot as pl
import warnings
warnings.filterwarnings('ignore')
```

```python
V = [0.0625, 0.125, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0]
data = np.loadtxt("data.csv", delimiter=",")

TIME   = [data[4*i].tolist() for i in range(10)]
BETAV  = [np.abs(data[4*i + 1]).tolist() for i in range(10)]
ALPHAX = [data[4*i + 2].tolist() for i in range(10)]
ALPHAV = [data[4*i + 3].tolist() for i in range(10)]

INV_TIME = [[1/t for t in TIME[i]] for i in range(10)]

print(ALPHAX[0])
```

```python
# RICHARDSON EXTRAPOLATION
fit_bv = [np.polyfit(INV_TIME[i], BETAV[i], 2, rcond=None, full=False, w=None, cov=False) for i in range(10)]
bv = [fit_bv[i][-1] for i in range(10)]

fit_ax = [np.polyfit(INV_TIME[i][3:], ALPHAX[i][3:], 2, rcond=None, full=False, w=None, cov=False) for i in range(10)]
ax = [fit_ax[i][-1] for i in range(10)]

fit_av = [np.polyfit(INV_TIME[i], ALPHAV[i], 2, rcond=None, full=False, w=None, cov=False) for i in range(10)]
av = [fit_av[i][-1] for i in range(10)]
```

```python
X = np.linspace(0.0, 0.035, num=1000)
Y_bv = [fit_bv[i][0]*X**2 + fit_bv[i][1]*X + fit_bv[i][2] for i in range(10)]
Y_ax = [fit_ax[i][0]*X**2 + fit_ax[i][1]*X + fit_ax[i][2] for i in range(10)]
Y_av = [fit_av[i][0]*X**2 + fit_av[i][1]*X + fit_av[i][2] for i in range(10)]
```

```python
epsilon = 0.2
t0 = 0.0
x0 = 300.0
v = 2.0
```

```python
ls = 2**13
data = np.fromfile('soliton.bin')
lt = np.shape(data)[0]//(3*ls)
data = data.reshape(np.shape(data)[0]//3, 3)
space = np.zeros((ls))
time = np.zeros((lt))
phi = np.zeros((lt,ls))
for i in range ((lt)):
    time[i] = data[ls*i,0]
    
for i in range ((ls)):
    space[i] = data[i,1]
    
for i in range ((lt)):
    for j in range ((ls)):
        phi[i,j] = data[ls*i+j,2]
```

```python
time
```

```python
# Renormalization
C = 1/(v + epsilon*(fit_alphav[0]*v**3 + fit_alphav[1]*v**2 + fit_alphav[2]*v + fit_alphav[3]))**2
def vR(t):
    return 1/np.sqrt(2*epsilon*fit_betav[0]*t + C)
    
xR0 = x0 + epsilon*(fit_alphax[0]*v**3 + fit_alphax[1]*v**2 + fit_alphax[2]*v + fit_alphax[3])
def xR(t):
    return (np.sqrt(2*epsilon*fit_betav[0]*t + C) - np.sqrt(C))/(epsilon*fit_betav[0]) + xR0

def u(t, x):
    return (np.sqrt(vR(t))/2) * (x - xR(t))

def ren_soliton(t, x):
    return (vR(t)/2) * np.cosh(u(t, x))**(-2)
```

```python
for i in range(1):
    # pl.plot(space, phi[i,:], color='red')
    # pl.plot(space, ren_soliton(time[i], space), color='blue')
    pl.plot(space, phi[i,:] - ren_soliton(time[i], space), color='blue')
pl.xlim(250.0, 450.0)
```

```python
from matplotlib.animation import FuncAnimation


fig, ax = pl.subplots()
xdata, ydata = [], []
xdata2,ydata2= [], []
ln, = pl.plot([], [], '-k', label='exact') 
ln2, = pl.plot([], [], '-b', label='renormalized')
pl.legend()

def init():
    ax.set_xlim(275, 350)
    ax.set_ylim(-0.2, 1.2)
    ln.set_data([], [])
    ln2.set_data([], [])
    return ln,ln2

def animate(i):
    ln.set_data(space, phi[i,:] )
    
    ln2.set_data(space, ren_soliton(time[i], space))
    return ln,ln2

anim = FuncAnimation(fig, animate, init_func=init, frames=len(time), blit=True)
anim.save('ren.mp4', writer='ffmpeg', fps=200)
```
